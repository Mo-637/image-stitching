```python
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
img1_path = os.path.join(os.path.abspath(''), 'adobe_panoramas', 'data', 'hotel', 'hotel-00.png')
assert os.path.exists(img1_path) == 1
img2_path = os.path.join(os.path.abspath(''), 'adobe_panoramas', 'data', 'hotel', 'hotel-01.png')
assert os.path.exists(img2_path) == 1
```

```python
img_1 = cv2.imread(img1_path)
img_2 = cv2.imread(img2_path)
```

```python
sift = cv2.SIFT_create()

keypoints1, descriptor1 = sift.detectAndCompute(img_1, None)
keypoints2, descriptor2 = sift.detectAndCompute(img_2, None)

keypoint_image1 = img_1[:];
cv2.drawKeypoints(img_1, keypoints1, keypoint_image1);

keypoint_image2 = img_2[:];
cv2.drawKeypoints(img_2, keypoints2, keypoint_image2);
```

```python
fig = plt.figure(figsize=(20,10))

ax = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(keypoint_image1)
ax.set_title('Image 0')

ax = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(keypoint_image2)
ax.set_title('Image 1')

plt.show()
```
